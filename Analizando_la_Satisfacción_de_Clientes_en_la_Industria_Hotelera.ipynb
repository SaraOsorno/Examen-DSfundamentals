```python
df['company'] = df['company'].fillna(df['agent'])
df[['company', 'agent']].head()
```

**Nota**: De acuerdo a lo observado en el csv en kaggle sobre la definicion de las columnas de 'company' y 'agent' se tienen la siguiente definicion:
**agent:ID of the travel agency that made the booking**
**Company: ID of the company/entity that made the booking or responsible for paying the booking. **

Dado esto el ID DEL AGENT Y DEL COMPANY se podria decir que son similares , por lo que use el comando fillna() para llenar los valores nulos de 'country' con los valores de 'agent'

```python
df.info()
```

```python
df.describe().T
```

**Nota**: De acuerdo a lo observado en el describe , se ve un valor atipico el cual significa 'stays_in_week_nights'Número de noches entre semana (de lunes a viernes) que el huésped se alojó o reservó para alojarse en el hotel. Teniendo un valor maximo de 41 lo cual no es posible ya que el maximo de noches que un huesped se puede alojar de lunes a viernes es 5.

```python
filtro_4 = df[df['stays_in_week_nights'] > 5]
cantidad = filtro_4.shape[0]
cantidad
```

**Nota:** Como se observa realizamos un fitro con los valores que eran mayor a 5 y se filtro , para luego usar el comando .shape[0] para mirar la cantidad de valores que eran mayores a este numero

```python
df.loc[df['stays_in_week_nights'] > 5, 'stays_in_week_nights'] = 5
df
```

**Nota**: Aca simplemento lo que hice fue usar el metofo loc, en el que le especfique cierta condicion y reemplazarla por un numero que yo indique. En este caso 5 que es el maximo de noches que hay de lunes  a viernes

Aca haremos lo mismo para la variable 'stays_in_weekend_nights' aca tambien se ve un valor atipico como maximo 16 , cuando por definicion Número de noches de fin de semana (sábado o domingo) que el huésped se alojó o reservó para alojarse en el hotel. Por lo que el valor maximo deberia ser 2

```python
df.loc[df['stays_in_weekend_nights'] > 2, 'stays_in_weekend_nights'] = 2
df
```

```python
df.describe().T
```

**Nota**: Como vemos el maximo ya seria 5 Y 2
